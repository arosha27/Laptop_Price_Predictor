```python
# Load the data 
import os
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#load engineered data for EDA
v2_data = pd.read_csv("../Data/Processed/v2_engineered.csv")
```

```python
v2_data.columns
```

```python
#set asethetic style 
sns.set(style="whitegrid")
```

### 1. Price Distribution

```python
## 1. Price Distribution
plt.figure(figsize=(8,4))
sns.histplot(v2_data["Price"] , kde=True , color = "blue")
plt.title("Price Distribution")
plt.savefig("../Data/Assets/price_distribution.png")
plt.show()
```

### Company Vs Price

```python
# 2. Company vs Price
plt.figure(figsize=(8,4))
sns.barplot(x=v2_data['Company'], y=v2_data['Price'] , color = "orange")
plt.xticks(rotation='vertical')
plt.title('Average Price by Company')
plt.savefig('../Data/Assets/company_vs_price.png')
```

### 3. TypeName vs Price


```python
plt.figure(figsize=(8,4))
sns.barplot(x=v2_data['TypeName'], y=v2_data['Price'])
plt.xticks(rotation='vertical')
plt.title('Average Price by Laptop Type')
plt.savefig('../Data/Assets/type_vs_price.png')
```

### 4. Ram vs Price

```python
plt.figure(figsize=(8,4))
sns.barplot(x=v2_data['Ram'], y=v2_data['Price'])
plt.title('Ram vs Price')
plt.savefig('../Data/Assets/ram_vs_price.png')
```

### 5. OS vs Price


```python
plt.figure(figsize=(8,4))
sns.boxplot(x=v2_data['os'], y=v2_data['Price'])
plt.title('Operating System vs Price')
plt.savefig('../Data/Assets/os_vs_price.png')
```

### 6. Correlation Heatmap


```python
plt.figure(figsize=(12 , 6))
# Select only numeric columns for correlation
numeric_df = v2_data.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.savefig('../Data/Assets/correlation_heatmap.png')
```

### 7. Pairplot of numerical features

```python
# 7. Pairplot of numerical features
sns.pairplot(v2_data[['Price', 'Ram', 'Weight', 'ppi', 'SSD', 'HDD']])
plt.savefig('../Data/Assets/pairplot.png')
```
